```python
import bs4
```

```python
import requests
import pandas as pd
```

```python
item = 1
name_list = []
price_list =[]
while item <= 32:
  data = requests.get('https://www.bigtone.in.th/product-category/electric/page/'+str(item))
  soup = bs4.BeautifulSoup(data.text)
  print(len(soup.find_all("li",{"class" :"product"})))
  for c in soup.find_all("li",{"class" :"product"}):
        guitar = c.find("div",{"class" :"product-details"})
        name_list.append(guitar.find('a').text)

        c.find("span",{"class" :"price"})
        price_list.append(c.find('bdi').text.strip("฿\xa0"))



  print("success page",(item))
  item +=1


print(name_list)
print(price_list)
table = pd.DataFrame([name_list,price_list]).transpose()
table.columns = ['guitar_name','price']

table.index +=1

table.to_excel('bigtone guitar all.xlsx',engine='openpyxl')
```
